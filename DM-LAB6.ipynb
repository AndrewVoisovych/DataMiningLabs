```python
### Code starts from here ###
import pandas as pd

url = 'https://raw.githubusercontent.com/AndrewVoisovych/DataMiningLabs/master/understat.com.csv'
dataset = pd.read_csv(url)

dataset = dataset[['сhampionship', 'season','matches','position', 'team', 'wins', 'loses', 'draws', 'scored', 'xG', 'deep', 'deep_allowed' , 'xG_diff', 'missed', 'xGA','pts','xpts', 'xpts_diff']]
dataset = dataset.drop(dataset[dataset['сhampionship'] == 'RFPL'].index)

dataset.head(10)
```

###2.Оберіть метрику яку ви використаєте для ваших даних і задачі  та аргументуйте чому обрали її 


Для порівняння двох алгоритмів, я обрав метрику Mean Absolute Error (MAE) для того, щоб подивитися абсолютну помилку при прогнозуванні. Середня абсолютна помилка являє собою суму абсолютних відмінностей між прогнозами і фактичними значеннями. Це дає уявлення про те, наскільки неправильними були прогнози.

mean squared error -  Квадратний корінь середнього значення помилок у квадраті. Чим менше тим краще, обраховується за допомогою співвідношення до середного по стовпці.

Функція max_error обчислює максимальну залишкову помилку - метрику, яка фіксує найгіршу помилку між прогнозованим значенням та істинним значенням. В ідеально підібраній моделі регресії вихідне значення max_error буде 0 на навчальному наборі, і хоча це буде малоймовірно в реальному світі, цей показник показує ступінь помилки, яку мала модель, коли вона була навчена.

Метрика R^2 вказує наскільки точно набір прогнозів відповідає фактичним значенням. Значення 0 - дуже погана відповідність, значення 1 - ідеальна. Негативна оцінка може бути в разі дуже поганої моделі.

Такі ж метрики, як Mean Poisson, Gamma, and Tweedie deviances не можуть бути використані, через те що забагато колонок йде на навчання.  Mean squared logarithmic error метрика, не скаже нічого нового, через те що вже присутні подібні. А Explained variance score - всього лиш не зовсім коректний R^2. 


###3.Використайте вашу обрану метрику на натренованій моделі на тестових даних. Наведіть результати метрики для моделей з 4 і 5 лабораторної
 

```python
%matplotlib inline
from sklearn import metrics
import numpy as np  
from sklearn.model_selection import train_test_split 
import warnings
warnings.filterwarnings('ignore')

X = dataset[['deep', 'deep_allowed', 'xG']]
X = pd.get_dummies(data = X)
y = dataset['pts'].values

X_train, X_test, y_train, y_test = train_test_split(X.values, y, test_size=0.2, random_state=0)
```

```python
from sklearn import linear_model
regr1 = linear_model.ARDRegression()
regr1.fit(X_train, y_train)
predicted1 = regr1.predict(X_test)


ARD_MSE =  np.sqrt(metrics.mean_squared_error(y_test, predicted1))
ARD_MAE = metrics.mean_absolute_error(y_test, predicted1)
ARD_ME = metrics.max_error(y_test, predicted1)
ARD_R2 = metrics.r2_score(y_test, predicted1)

print("ARDRegression\n")
print('Квадратний корінь середнього значення помилок у квадраті:\n ', ARD_MSE)
print('Чим менше значення , тим краще')

print("\nСередня абсолютна помилка.")
print('МАЕ: ', ARD_MAE)
print('Значення 0 вказує на відсутність помилок тобто те, що прогноз ідеальний')

print("\nМаксимальна залишкова помилка")
print('max_error: ', ARD_ME)

print('\nКоефіцієнт детермінації  r^2: ', ARD_R2)
print('Вказує наскільки отримані спостереження підтверджують модель.')
      
```

```python
regr2 = linear_model.LinearRegression()
regr2.fit(X_train, y_train)
predicted2 = regr2.predict(X_test)


L_MSE =  np.sqrt(metrics.mean_squared_error(y_test, predicted2))
L_MAE = metrics.mean_absolute_error(y_test, predicted2)
L_ME = metrics.max_error(y_test, predicted2)
L_R2 = metrics.r2_score(y_test, predicted2)

print("LinearRegression\n")
print('Квадратний корінь середнього значення помилок у квадраті:\n ', L_MSE)
print('Чим менше значення , тим краще')

print("\nСередня абсолютна помилка.")
print('МАЕ: ', L_MAE)
print('Значення 0 вказує на відсутність помилок тобто те, що прогноз ідеальний')

print("\nМаксимальна залишкова помилка")
print('max_error: ', L_ME)

print('\nКоефіцієнт детермінації  r^2: ', L_R2)
print('Вказує наскільки отримані спостереження підтверджують модель.')

      
```

```python
import pandas as pd

dMSE = L_MSE - ARD_MSE; 
dMAE = L_MAE - ARD_MAE;
dME = L_ME - ARD_ME;
dR2 = ARD_R2 - L_R2;

df = {
    'Regression': ['ARD', 'Linear', 'Diff'],
    'MSE': [ARD_MSE, L_MSE, dMSE],
    'МАЕ': [ARD_MAE, L_MAE, dMAE],
    'max_error': [ARD_ME, L_ME, dME],
    'r^2': [ARD_R2, L_R2, dR2]
    }
print(pd.DataFrame(data=df))

```
